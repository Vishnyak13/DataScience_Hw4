## Перша частина

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
```

Є набір даних speed, який представляє значення швидкості для деякого транспортного засобу в певний момент спостереження. Очевидно, що дані мають дискретний вигляд. Відомо, що спостереження відбувалися з періодом в одну годину.

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

1. Приймемо, що в нульовій координаті у нас швидкість 25 км/год, відповідно до набору даних speed. Помістіть у змінну time - вектор часу, створений за допомогою np.linspace (всього 12 спостережень, від 0 до 11 годин)

```python
time = np.linspace(0, 11, 12)
```

2. Виконайте виведення масиву time

```python
time
```

3. Виконайте виведення графіка — значення точок швидкості (plot або scatter). Задайте розмір області, що відображається (0, 11) і (0, 130). Задайте відображення сітки

```python
plt.figure(figsize=(16, 9))
plt.plot(time, speed, 'bo', label='Точки швидкості')
plt.plot(time, speed, 'r--')
plt.legend()
plt.grid()
plt.title('Значення точок швидкості')
plt.ylabel('Швидкість, км/год')
plt.xlabel('Час, год');
```

4. Виконайте інтерполяцію за допомогою interp1d(kind='cubic') та отримайте функцію на 10000 значень. Побудуйте безперервний графік отриманої функції.

```python
f = interp1d(time, speed, kind='cubic')
ext_time = np.linspace(0, 11, 10000)
plt.figure(figsize=(16, 9))
plt.plot(time, speed, 'bo', label='Точки швидкості')
plt.plot(ext_time, f(ext_time), 'g--')
plt.legend()
plt.grid()
plt.title("'Графік безперервної функції за умови kind='cubic'")
plt.ylabel('Швидкість, км/год')
plt.xlabel('Час, год');
```

5. Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]

```python
distance_cubic = integrate.quad(f, 0, 11)
distance_cubic
```

6. Виконайте також пункти 4 та 5 для kind='quadratic'

```python
f_quad = interp1d(time, speed, kind='quadratic')
ext_time = np.linspace(0, 11, 10000)
plt.figure(figsize=(16, 9))
plt.plot(time, speed, 'bo', label='Точки швидкості')
plt.plot(ext_time, f(ext_time), 'g--', label='cubic')
plt.plot(ext_time, f_quad(ext_time), 'r-', label='quadratic', linewidth=2)
plt.legend()
plt.grid()
plt.title("Графік безперервної функції")
plt.ylabel('Швидкість, км/год')
plt.xlabel('Час, год');
```

```python
distance_quad = integrate.quad(f_quad, 0, 11)
distance_quad
```
